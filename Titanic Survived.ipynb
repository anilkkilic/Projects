```python
import pandas as pd
import numpy as np
```

```python
from IPython.display import display
data_train = pd.read_csv('train (1).csv')
display(data_train.head())
data_test = pd.read_csv('test (1).csv')
display(data_test.head())
```

# Embarked

```python
embarked_mapping = {"S": 1, "C": 2, "Q": 3}
data_train['Embarked'] = data_train['Embarked'].map(embarked_mapping)
data_test['Embarked'] = data_test['Embarked'].map(embarked_mapping)
```

# Sex

```python
sex_mapping = {"male": 0, "female": 1}
data_train['Sex'] = data_train['Sex'].map(sex_mapping)
data_test['Sex'] = data_test['Sex'].map(sex_mapping)
```

# Cabin
Cabin column has 148 unique values. Taking first letter of cabin values and mapping them.

```python
unique = data_train['Cabin']
print(unique.unique())
```

```python
type(unique)
a = np.array(unique.values.tolist())
firsts = [w[0].lower() for w in a]
cabin_frame = pd.DataFrame(firsts, columns=['Cabin_mapped'])
cabin_frame
```

```python
cabin_mapping = {"n": 0, "a": 1, "b": 2, "c": 3, "d": 4, "e": 5, "f": 6, "g":7}
cabin_frame['Cabin_mapped'] = cabin_frame['Cabin_mapped'].map(cabin_mapping)
```

```python
data_train = data_train.assign(Cabin_mapped = cabin_frame['Cabin_mapped'])
data_train
```

# Drop
We don't need cabin and ticket columns.

```python
data_train = data_train.drop(['Cabin','Ticket'], axis=1)
data_train
```

# Fill NA
Filling NaN columns with mean.

```python
print(pd.isnull(data_train).sum())
```

```python
mean_age_data_train = data_train['Age'].mean()
data_train['Age'] = data_train['Age'].fillna(mean_age_data_train)
mean_cabin_mapped_data_train = data_train['Cabin_mapped'].mean()
data_train['Cabin_mapped'] = data_train['Cabin_mapped'].fillna(mean_cabin_mapped_data_train)
mean_embarked_data_train = data_train['Embarked'].mean()
data_train['Embarked'] = data_train['Embarked'].fillna(mean_embarked_data_train)
print(pd.isnull(data_train).sum())
```

# Name
Rev. Mr. Mrs. Miss. Master. Don. Dr. Mme. Ms. Major. Lady. Sir. Col. Mlle. Capt. Hepsi nokta içeriyor. Ona göre method.

```python
# pd.options.display.max_rows = 1000
# display(data_train)
data_train = data_train.drop(['Name'], axis=1)
data_test = data_test.drop(['Name'], axis=1)
```

# Age
To increase the accuracy normalize the age.

```python
x_normed = data_train['Age'] / data_train['Age'].max(axis=0)
data_train['Age_mapped'] = data_train['Age'].map(x_normed)
data_train = data_train.drop(['Age'], axis=1)
mean_age = data_train['Age_mapped'].mean()
data_train['Age_mapped'] = data_train['Age_mapped'].fillna(mean_embarked_data_train)
data_train
```

# Fare

```python
y_normed = data_train['Fare'] / data_train['Fare'].max(axis=0)
data_train['Fare_mapped'] = data_train['Fare'].map(y_normed)
mean_fare = data_train['Fare_mapped'].mean()
data_train['Fare_mapped'] = data_train['Fare_mapped'].fillna(mean_fare)
data_train = data_train.drop(['Fare'], axis=1)
data_train
```

```python
from sklearn.model_selection import train_test_split

predictors = data_train.drop(['Survived', 'PassengerId'], axis=1)
target = data_train["Survived"]
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.25, random_state = 0)
predictors
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
acc_logreg = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_logreg)
```

```python
from sklearn.svm import LinearSVC

linear_svc = LinearSVC()
linear_svc.fit(X_train, y_train)
y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_linear_svc)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
acc_knn = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_knn)
```

```python
tests = pd.DataFrame({
    'Model': ['Logistic Regression', 'LinearSVC', 'KNN'],
    'Accuracy': [acc_logreg, acc_linear_svc, acc_knn]
    })
tests
```
